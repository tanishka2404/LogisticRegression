```python
#import sklearn library to train the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(dia.drop("Outcome",axis=1),dia['Outcome'],test_size=0.3)
#x_train is training set of indepedent variable
#x_test is test set of independent variable
#y_train is training set of dependent variable
#y_test is test set of independent variable
```

```python
#now import linear model i.e, logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis_regr=LogisticRegression()
#we created instance here
```

```python
logis_regr.fit(x_train,y_train)
#we are fitting model on dataset
```

```python
y_predict=logis_regr.predict(x_test)
```

```python
y_test.head()
```

```python
y_predict[0:5]
```

```python
#you can see the difference in the actual value and predicted value
```

```python
#now use sklearn metrices and import classificaton report
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_predict)
```
